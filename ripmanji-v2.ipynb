```python
# importing the necessary libraries
import numpy as np
import pandas as pd
```

```python
# importing the dataframe 
df = pd.read_csv('kan_text_516400.csv', header=None, names=['Texts', 'Sentiment'])
df.head()
```

```python
kan_text = df['Texts']
label = df['Sentiment']
```

```python
# tokenizing the text
'''
Sentence: "He is a good boy"
After tokenizing: ['He', 'is', 'a', 'good', 'boy']
'''
lst1 = []
for i in kan_text:
    temp = i.split(' ')
    lst1.append(temp)
```

```python
# finding out the length of the longest piece of text
longest = 0
for i in kan_text:
    if len(i) > longest:
        longest = len(i)
        sentence = i

print(longest)
```

```python
# assigning each word an integer
'''
An integer is assigned to all the unique words in the corpus
'''
counter = 0
dictionary = {}
for i in kan_text:
    for j in i:
        if j not in dictionary:
            dictionary[j] = counter 
            counter = counter + 1
```

```python
# text to sequences
lst2 = []
for i in kan_text:
    temp = []
    for j in i:
        temp.append(dictionary[j])
    lst2.append(temp)
```

```python
# padded the text to sequences to make them all the same length 
lst3 = []
for i in lst2:
    temp = longest - len(i)
    i = i + [0]*temp
    lst3.append(i)

# converting list to numpy array
lst4 = np.asarray(lst3, dtype=np.float64)
```

```python
X = lst4
y = label
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# clf = LogisticRegression().fit(X_train, y_train)
clf = RandomForestClassifier().fit(X_train, y_train)
```

```python
# passing the X_test values and seeing their predictions
y_pred = clf.predict(X_test)

# comparing the predictions with the actual y_test values
(y_pred==y_test).all()
```

```python
# Return the mean accuracy on the given test data and labels.
clf.score(X_test, y_test)
```
